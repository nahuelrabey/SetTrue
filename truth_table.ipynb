```python
import pandas as pd
import numpy as np

def gen_prepositions_var(prep_qtty: int, names:'list[str]' = None)-> pd.DataFrame:
    # preps = []
    if (names and (len(names) != prep_qtty)):
        raise TypeError("names length should be equal to prepositions quantity")
    posible_valuess = 2**prep_qtty

    data_dic = {}
    for x in range(prep_qtty):
        col = []
        variator = 2**(x+1)
        series_length = int(posible_valuess/variator)
        prep_value = True
        # how many series are for current prep.
        series_noumber = int(posible_valuess/series_length)
        for serie in range(series_noumber):
            # print("run:", run)
            # print("step:", [int(step) for x in range(series_length)])
            col = col + [prep_value for x in range(series_length)]
            prep_value = not prep_value

        # print("col:", col)
        name = names[x] if names else x
        data_dic[name] = col
    
    # return data_dic
    return pd.DataFrame(data_dic)


# def xor(a: bool,b: bool):
#     return (a or b) and not (a and b)

# df = gen_prepositions_var(2, ["p","q"] )
# df["p and q"] = df["p"] & df["q"]
# df["p or q"] = df["p"] | df["q"]
# df
```
